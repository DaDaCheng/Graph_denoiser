```python
import numpy as np
from WPGD import WPGD
import torch
import networkx as nx
```

```python
G=nx.generators.lattice.grid_2d_graph(5,5)
A_target=nx.adjacency_matrix(G).toarray()
nx.draw(G)
```

```python
n=10
N=n*n
G2=nx.generators.lattice.grid_2d_graph(n,n)
A2=nx.adjacency_matrix(G2).toarray()
nx.draw(G2)
```

```python
# add noise
NA=torch.ones(N,N)*(20/N**2)
NA=torch.bernoulli(NA).clone().numpy()
NA=NA+A2
NA=NA+NA.T
NA[NA>0]=1
```

```python
G3=nx.from_numpy_matrix(NA)
nx.draw(G3)
```

```python
Graph_Denoiser=WPGD()
Graph_Denoiser.target_graph(A_target)
A_denoised=Graph_Denoisor.denoise_sync(NA,thre=-15)
```

```python
G4=nx.from_numpy_matrix(A_denoised)
nx.draw(G4)
```
